```python
import pandas as pd
import numpy as np
import warnings as w
w.filterwarnings ('ignore')
```

```python
# /here i read two files
sheet1=pd.read_excel('Downloads/campaign/Campaign data.xlsx',sheet_name=0)
sheet2=pd.read_excel('Downloads/campaign/Campaign data.xlsx',sheet_name=1)
```

```python
sheet1
```

```python
# to drop duplicates
sheet1[sheet1.duplicated()]
sheet1.drop_duplicates(inplace=True)
```

```python
# HERE IM MERGING THE SHEET1 AND sheet2 by using (.merg)
sheet=pd.merge(sheet1,sheet2)
```

```python
# converting the date column to separate date,month,year
sheet['Date']=pd.to_datetime(sheet['Date'])
sheet['Day']=sheet['Date'].dt.day
sheet['Month']=sheet['Date'].dt.month
sheet['Year']=sheet['Date'].dt.year
```

```python
sheet.drop_duplicates(inplace=True)
```

```python
sheetinfo=sheet.info()
sheetisnul=sheet.isnull().sum()
sheetshape=sheet.shape
sheetdtype=sheet.dtypes
sheetcolumns=sheet.columns
sheetdesc=sheet.describe()
sheethead=sheet.head()
sheettail=sheet.tail()
sheetsample=sheet.sample()
```

```python
#1 Find out the top 5 months in terms of impressions
# k=sheet.groupby(['Month','Year','Metric'])['Value'].sum().reset_index()
imp=sheet[sheet['Metric']=='Impressions']
maximp=imp.groupby(['Month','Year'])['Value'].sum().sort_values(ascending=False)[:5]
maximp
```

```python
#2 Which model has maximum impressions?
h=imp.groupby(['Model'])['Value'].sum().sort_values(ascending=False)[:1]
h
```

```python
#3 Find out the top 3 campaigns in terms of impressions for every message type
imp.groupby(['Campaign','Message type'])['Value'].sum().sort_values(ascending=False)[:3]
```

```python
# 5Find out the top 3 campaigns in terms of CPM (CPM stands for cost per mile which means cost per 1000 impressions. CPM = total media cost/total impressions * 1000)
media=sheet[sheet['Metric']=='Media Cost']
imprs=sheet[sheet['Metric']=='Impressions']
```

```python

med=media.groupby(['Metric','Campaign'])['Value'].sum().reset_index()
impi=imprs.groupby(['Metric','Campaign'])['Value'].sum().reset_index()
```

```python
merged=pd.merge(med,impi,on='Campaign')
```

```python
# calculating the Cpm
merged['CPM']=(merged['Value_x'])/ (merged['Value_y'])*1000
merged
```
